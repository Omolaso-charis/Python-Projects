Using the Ecommerce purchases dataset, answer the questions below

import pandas and read in the Ecommerce Purchases csv file and set it to a dataframe called df

```python
df = pd.read_csv("Ecommerce_Purchases.csv")
df
```

check the head of the dataframe

How many rows and columns are there

```python
df.head(0)
```

```python
rows = df.shape[0]
columns = df.shape[1]
print("number of rows in df:" , rows)
print("number of columns in df:" , columns)
```

What is the average purchase price?

```python
average_price = df["Purchase Price"].mean()
average_price
```

What were the highest and lowest purchase prices?

```python
max_price = df["Purchase Price"].max()
min_price = df["Purchase Price"].min()
print("max price :" , max_price)
print("min price :" , min_price)
```

How many people have English "en" as their language of choice on the website?

```python

print(df['Language'].value_counts()["en"])
```

How many people have the job title of "Lawyer"?

```python
print(df['Job'].value_counts()["Lawyer"])
```

How many people made the purchase during the AM and how many people made the purchaseduring the PM

```python
AM = df["AM or PM"].value_counts()["AM"]
PM = df["AM or PM"].value_counts()["PM"]
print("AM" , AM)
print("PM" , PM)
```

What are the 5 most common job titles?

```python
sort = df['Job'].value_counts().head(5)
sort
```

someone made a purchase that came from Lot: "90 WT", what was the purchase price for this transaction?

```python
purchase_price = df[df['Lot'] =="90 WT"]['Purchase Price']
purchase_price
```

What is the email of the person with the following credit card number: 4926535242672853

```python
email = df[df['Credit Card'] == 4926535242672853]['Email']
email
```

How many people have American express as their credit card provider and made a purchase above $95?

```python
df1= df.query("`Purchase Price` > 95")

print(df1['CC Provider'].value_counts()["American Express"])
```

How many people have a crdit card that expires in 2025?

```python

df[df['CC Exp Date'].str[3:] == "25"].count()[0]
```

What are the top 5 most popular email providers/hosts(e.g.gmail.com,yahoo.com,etc)

```python
df['Email prov'] = df["Email"].str.split("@", n = 1, expand = True)[1]
result = df.groupby('Email prov')['Email prov'].count().nlargest(5)
result
```

The duration column contains data on the film length. How many missing values are there in this column?

```python
df2["duration"].isnull().sum()
```

Replace the missing values in the duration column with the median value for this column.

```python
df3 = df2["duration"].fillna(value = df2["duration"].median())
df3
```

What is the average film length? Give the answer as a floating-point figure rounded to two decimal places.
